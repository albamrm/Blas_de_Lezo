```python
import random

class Tablero:
    def __init__(self):
        self.tablero = [['^' for _ in range(10)] for _ in range(10)]
        self.tablero_disparos = [['^' for _ in range(10)] for _ in range(10)]

    def imprimir_tablero(self):
        print("  A B C D E F G H I J")
        for i in range(10):
            print(i, end=' ')
            for j in range(10):
                print(self.tablero[i][j], end=' ')
            print()

    def imprimir_tablero_disparos(self):
        print("  A B C D E F G H I J")
        for i in range(10):
            print(i, end=' ')
            for j in range(10):
                print(self.tablero_disparos[i][j], end=' ')
            print()

    def colocar_barcos(self):
        barcos = [4, 3, 3, 2, 2, 2, 1, 1, 1, 1]
        direcciones = [(0, 1), (1, 0)]
        for tamaño_barco in barcos:
            while True:
                x = random.randint(0, 9)
                y = random.randint(0, 9)
                dirección = random.choice(direcciones)
                válido = True
                for i in range(tamaño_barco):
                    nuevo_x = x + dirección[0] * i
                    nuevo_y = y + dirección[1] * i
                    if not (0 <= nuevo_x < 10 and 0 <= nuevo_y < 10) or self.tablero[nuevo_x][nuevo_y] != '^':
                        válido = False
                        break
                if válido:
                    for i in range(tamaño_barco):
                        self.tablero[x + dirección[0] * i][y + dirección[1] * i] = str(tamaño_barco)
                    break

# Crear un objeto Tablero
tablero_jugador = Tablero()

# Colocar barcos en el tablero del jugador
tablero_jugador.colocar_barcos()

# Mostrar tablero del jugador y tablero de disparos
print("Tablero del jugador:")
tablero_jugador.imprimir_tablero()
print("\nTablero de disparos:")
tablero_jugador.imprimir_tablero_disparos()
        
```
